```python
import pandas as pd
import os
import numpy as np
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
```

```python
results_path = '/nas/home/cborrelli/tripletloss_bot/results'

df_dev = pd.read_pickle(os.path.join(results_path, 'dev_results.pkl'))



labels = np.where(df_dev['class_1']==df_dev['class_2'], 0, 1)
df_dev['labels'] = labels

dev_classes_list = ['-', 'A01', 'A02', 'A03', 'A04', 'A05', 'A06']


for c in dev_classes_list:  
    df_part = df_dev[((df_dev['class_1'] == '-')&(df_dev['class_2'] == '-')) |
                    ((df_dev['class_1'] == '-')&(df_dev['class_2'] == c)) |
                     ((df_dev['class_1'] == c)&(df_dev['class_2'] == '-'))
                    ]
    fpr, tpr, thresholds = roc_curve(df_part['labels'], df_part['distance'])
    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='red',
             lw=lw)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC for class {}'.format(c))
    plt.legend(loc="lower right")
    plt.show()
```
